```python
import math
def GetCommonNeighbors(u, v):
    try:
        with open("C:/Users/Cheryl Zhang/Desktop/train.txt") as trainfile:
            for i, line in enumerate(trainfile):
                line_list = [int(k) for k in line[:-1].split("\t")];
                if int(line_list[0]) == u:
                    usinksList = line_list[1:]
                    #print(usinksList)         
                    break
            for i, line in enumerate(trainfile):
                line_list = [int(k) for k in line[:-1].split("\t")];
                if int(line_list[0]) == v:
                    vsinksList = line_list[1:]
                    #print(vsinksList)         
                    break
            u_neighbors = set(usinksList)
            v_neighbors = set(vsinksList)
            return len(u_neighbors.intersection(v_neighbors))
    except:
        return 0

def AdamicAdar(u, v):
    try:
        with open("C:/Users/Cheryl Zhang/Desktop/train.txt") as trainfile:
            for i, line in enumerate(trainfile):
                line_list = [int(k) for k in line[:-1].split("\t")];
                if int(line_list[0]) == u:
                    usinksList = line_list[1:]
                    #print(usinksList)         
                    break
            for i, line in enumerate(trainfile):
                line_list = [int(k) for k in line[:-1].split("\t")];
                if int(line_list[0]) == v:
                    vsinksList = line_list[1:]
                    #print(vsinksList)         
                    break
            u_neighbors = set(usinksList)
            v_neighbors = set(vsinksList)
            interset = u_neighbors.intersection(v_neighbors)
            aa = 0
            for i in interset:
                try:
                    aa += 1 / math.log(vertex_dict[i])
                except:
                    aa += 0
    except:
        aa = 0
    return aa
```
